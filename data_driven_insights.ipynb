  - int64

```python

#Analisar o nome das colunas para entender melhor a base.
print(df.columns)

#Analisar o tipo de dado de cada coluna.
print(df.dtypes)
```

#Extrair e mostrar categoria de produtos


```python

#Converter a coluna "PRODUCTLINE" em uma lista Python para trabalhar com as estruturas mais básicas.
lista = list(df['PRODUCTLINE'])

#Imprimir a lista inteira para visualizar todas as categorias presentes.
print(lista)

#Usar um loop "for" para imprimir as primeiras 5 categorias da lista, facilitando a visualização inicial dos dados.
for categoria in lista[:5]:
  print(categoria)
```

```python
#Criar dicionário que relaciona código do produto com a categoria. O zip junta os dois elementos em pares. O primeiro código com a primeira categoria, o segundo código com a segunda categoria e assim por diante. O dict transforma esse pares em um dicionário Python onde cada chave é um código do produto(PRODUCTCODE) e cada valor é a categoria correspondente(PRODUCTLINE).

dicionario_produtos = dict(zip(df['PRODUCTCODE'], df['PRODUCTLINE']))


## Loop para exibir os primeiros pares chave-valor do dicionário.
"""
- Utilizar o método ".items()" do dicionário "dicionario_produtos" para obter uma lista de tuplas, onde cada tupla representa um par chave-valor (código do produto e categoria).
- Aplicamos o fatiamento "[:3]" para selecionar somente os três primeiros pares dessa lista, facilitando a visualização inicial sem imprimir o dicionário completo.
- O loop "for" itera sobre cada um desses pares, atribuindo a chave do produto à variável "chave" e a categoria correspondente à variável "valor".
- Em cada iteração, a função "print()" exibe o par chave-valor formatado usando uma f-string, garantindo uma saída legível no formato: "chave: valor".
"""
for chave, valor in list(dicionario_produtos.items())[:3]:
  print(f'{chave}: {valor}')
```

```python
#Prática de tuplas utilizadando o Data Frame.
#Objetivo:criar uma tupla que contém os dados do primeiro cliente, unindo os valores dessas 3 colunas.

#Selecionar a primeira linha do Data Frame(índice 0).
linha0 = df.iloc[0]

#Criar uma tupla com três informações distintas dessa linha.
#Exemplo: nome do cliente, cidade e valor de vendas.
minha_tupla = (linha0['CUSTOMERNAME'], linha0['CITY'], linha0['SALES'])

#Imprimir o resultado da tupla.
print(minha_tupla)
```

```python
#Utilizando as estruturas condicionais.
#Filtrar e exibir os nomes dos clientes com vendas acima de 5000.

# Criei uma nova variável chamada 'clientes_acima_5000' para armazenar o resultado do filtro.
# A condição df['SALES'] > 5000 retorna uma Série booleana (True/False) para cada linha do DataFrame.
# Apenas as linhas em que a condição for True serão mantidas no novo DataFrame.
clientes_acima_5000 = df[df['SALES'] > 5000]

# Em seguida,incluí o print() para exibir somente os nomes dos clientes que atendem à condição.
# Ou seja, acessamos apenas a coluna 'CUSTOMERNAME' do DataFrame filtrado.
print(clientes_acima_5000['CUSTOMERNAME'])
```

```python
#Operações aritméticas e manipulação de dados.

# Criei uma nova coluna 'BONUS' no DataFrame, que representa 10% do valor da venda ('SALES').
# Multipliquei a coluna 'SALES' por 0.10 para calcular o bônus.
# Em seguida, usei o método `.round(2)` para arredondar os valores para duas casas decimais,
# facilitando a leitura e mantendo a precisão adequada para os valores monetários.
df['BONUS'] = (df['SALES'] * 0.10).round(2)

# Exibe as cinco primeiras linhas das colunas 'SALES' e 'BONUS' para verificar o cálculo.
print(df[['SALES', 'BONUS']].head())

# Exibe as cinco primeiras linhas do DataFrame completo para uma visão geral inicial.
df.head()
```

```python
#NumPy e arrays numéricos
# Criei um array NumPy a partir da coluna 'SALES' do DataFrame para facilitar operações numéricas eficientes.
array_sales = np.array(df['SALES'])

# Calculei a soma total de vendas.
total_sales = array_sales.sum()

# Calculei a média das vendas.
media_sales = array_sales.mean()

# Encontrei o valor máximo (maior venda).
max_sales = array_sales.max()

# Encontrei o valor mínimo (menor venda).
min_sales = array_sales.min()

# Calculei o desvio padrão, que indica a dispersão dos valores em relação à média.
desvio_sales = array_sales.std()

# Exibe os resultados formatados em reais, com separador de milhares e duas casas decimais.
print(f'Total de vendas: R$ {total_sales:,.2f}')
print(f'Média de vendas: R$ {media_sales:,.2f}')
print(f'Maior venda: R$ {max_sales:,.2f}')
print(f'Menor venda: R$ {min_sales:,.2f}')
print(f'Desvio padrão das vendas: R$ {desvio_sales:,.2f}')
```

```python
# Agrupei os dados por categoria de produto (PRODUCTLINE) e somei as vendas totais por categoria.
vendas_por_categoria = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

print('Receita total por categoria:')
print(vendas_por_categoria)
print('\n')

# Calculei a média de vendas por item, dentro de cada categoria.
media_por_categoria = df.groupby('PRODUCTLINE')['SALES'].mean().sort_values(ascending=False).round(2)

print(' Receita média por item, por categoria:')
print(media_por_categoria)
```

```python
# Juntei soma e média em um único DataFrame agrupado por categoria. Utilizei o agg para poder aplicaqr várias funções ao mesmo tempo em cada grupo.
resumo_categoria = df.groupby('PRODUCTLINE')['SALES'].agg(
    Total_Receita='sum',
    Media_Receita='mean'
).round(2).sort_values(by='Total_Receita', ascending=False)

print(" Resumo por categoria:")
print(resumo_categoria)
```

#Gráfico de linha
  - No gráfico de linhas é possível perceber que o produto que mais gera receita.
  -  A categoria que mais gerou receita é: Classic Cars com um valor bem expressivo.
  - A categoria que menos gerou receita é: Trains

```python


# Preparei os dados.
vendas_por_categoria = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

# Criei o gráfico de linha.
plt.figure(figsize=(10,10))
ax = vendas_por_categoria.plot(kind='line', color='green', marker='o')


# Adicionei o valor ao lado de cada ponto.
for i, valor in enumerate(vendas_por_categoria):
    ax.text(i, valor *1.02, f'R$ {valor:,.2f}', fontsize=7, va='bottom', ha='left', backgroundcolor='#d9f2d9', color='black')

# Títulos e legendas
plt.title('Receita Total por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Receita Total (R$)')

# Rotacionei os nomes no eixo x para melhor leitura.
plt.xticks(rotation=45)

# Removi os números do eixo y (oculta os valores e a legenda do eixo y)
plt.gca().yaxis.set_visible(False)

# Grade leve no eixo y para referência visual.
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#Gráfico de barras
  - Com o gráfico de barras conseguimos analisar a receita de vendas por cada  páis. É possível verificar também qual páis precisa focar esforços comerciais e de marketing.
  - País com maior receita de vendas: EUA
  - País com menor receita de vendas: IRELAND

  - Comentário sobre cada uma das atribuições:
        i, valor * 1.02,                  # Posição acima da barra (2% a mais que o valor real)
        f'R$ {valor:,.2f}',               # Formata valor como moeda
        fontsize=9,                       # Tamanho da fonte
        va='bottom',                      # Alinha verticalmente por baixo
        ha='center',                      # Alinha horizontalmente ao centro da barra
        backgroundcolor='#d9f2d9',        # Fundo do texto em verde claro
        color='black',                    # Cor do texto
        rotation=90,                      # Rotaciona o texto para caber melhor
        fontweight='bold'                 # Deixa o texto em negrito

```python

# Agrupei os dados por país e soma as vendas, ordenando do maior para o menor valor.
vendas_por_categoria = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)

# Defini o tamanho da figura do gráfico.
plt.figure(figsize=(15, 7))

# Criei o gráfico de barras com as receitas por país.
ax = vendas_por_categoria.plot(kind='bar', color='green')

# Adicionei os valores de receita acima de cada barra.
for i, valor in enumerate(vendas_por_categoria):
    ax.text(
        i, valor * 1.02,
        f'R$ {valor:,.2f}',
        fontsize=9,
        va='bottom',
        ha='center',
        backgroundcolor='#d9f2d9',
        color='black',
        rotation=90,
        fontweight='bold'
    )

# Defini o limite superior do eixo Y para evitar que o maior valor fique cortado.
plt.ylim(top=vendas_por_categoria.max() * 1.5)

# Título e rótulos dos eixos com formatação personalizada.
plt.title('Vendas por País', fontsize=17, fontweight='bold')
plt.xlabel('País', fontsize=16, fontweight='bold')
plt.ylabel('Receita R$', fontsize=12, fontweight='bold')

# Rotacionei os nomes dos países no eixo X para melhorar a visualização.
plt.xticks(rotation=45, ha='right', fontweight='bold')

# Exibe o gráfico.
plt.show()
```

#Gráfico de dispersão
  - Este gráfico permite visualizar se há uma correlação entre a quantidade de itens pedidos e a receita gerada.
    -  O parâmetro alpha=0.5 adiciona transparência aos pontos, facilitando a visualização em regiões com muitos dados sobrepostos.

```python
# Verifica se as colunas 'QUANTITYORDERED' e 'SALES' estão presentes no DataFrame.
if 'QUANTITYORDERED' in df.columns and 'SALES' in df.columns:

    # Cria um gráfico de dispersão (scatter plot) relacionando a quantidade pedida e o valor da venda.
    plt.scatter(df['QUANTITYORDERED'], df['SALES'], alpha=0.5)

    # Define o título do gráfico
    plt.title('Dispersão: Quantidade Pedida vs Receita')

    # Define os rótulos dos eixos
    plt.xlabel('Quantidade Pedida')
    plt.ylabel('Receita (Sales)')

    # Adiciona grade ao fundo para facilitar a leitura
    plt.grid()

    # Exibe o gráfico
    plt.show()
```
